````python
import requests
from bs4 import BeautifulSoup

url_base = "https://bv.fapesp.br/pt/pesquisa/buscador/?q2=(((aplicacao_exact%3A%22Pesquisa%20Inovativa%20em%20Pequenas%20Empresas%20(PIPE)%22%20%20AND%20nivel%3A3)%20OR%20aplicacao_exact%3A%22PAPPE-PIPE%20III%22)%20OR%20(convenio%3A%22FINEP%20-%20PAPPE-PIPE%20III%22))%20AND%20((auxilio%3A*%20AND%20situacao_exact%3A%22Conclu%C3%ADdos%22))&page={}&count=50"
total_pages = 7

resumos = []

for page_num in range(1, total_pages + 1):
    url = url_base.format(page_num)
    try:
        response = requests.get(url, verify=False)
        
        if response.status_code == 200:
            print(f"Solicitação para a página {page_num} bem-sucedida!")
            
            soup = BeautifulSoup(response.text, "html.parser")
            
            for elemento in soup.find_all(class_="content resumo"):
                resumo_parsed = elemento.p.text.strip(" \n")
                resumos.append(resumo_parsed)
        else:
            print(f"Falha na solicitação para a página {page_num}: {response.status_code}")
    except requests.exceptions.SSLError as e:
        print(f"Erro SSL ao fazer solicitação para a página {page_num}: {e}")

print(resumos)
````

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url_base = "https://bv.fapesp.br/pt/pesquisa/buscador/?q2=(((aplicacao_exact%3A%22Pesquisa%20Inovativa%20em%20Pequenas%20Empresas%20(PIPE)%22%20%20AND%20nivel%3A3)%20OR%20aplicacao_exact%3A%22PAPPE-PIPE%20III%22)%20OR%20(convenio%3A%22FINEP%20-%20PAPPE-PIPE%20III%22))%20AND%20((auxilio%3A*%20AND%20situacao_exact%3A%22Conclu%C3%ADdos%22))&page={}&count=50"
total_pages = 7

resumos = []
data = {}

for page_num in range(1, total_pages + 1):
    url = url_base.format(page_num)
    try:
        response = requests.get(url, verify=False)
        
        if response.status_code == 200:
            print(f"Solicitação para a página {page_num} bem-sucedida!")
            
            soup = BeautifulSoup(response.text, "html.parser")
            
            # Extrair resumos da página atual
            for elemento in soup.find_all(class_="content resumo"):
                resumo_parsed = elemento.p.text.strip(" \n")
                resumos.append(resumo_parsed)
            
            # Extrair dados da tabela da página atual
            for tabela in soup.find_all('table', id="processo_listagem"):
                for linha_tabela in tabela.find_all("tr"):
                    if linha_tabela.contents[0].text == 'CNAE:':
                        continue
                    if linha_tabela.contents[0].text not in data:
                        data[linha_tabela.contents[0].text] = []
                    data[linha_tabela.contents[0].text].append(linha_tabela.contents[1].text)
        else:
            print(f"Falha na solicitação para a página {page_num}: {response.status_code}")
    except requests.exceptions.SSLError as e:
        print(f"Erro SSL ao fazer solicitação para a página {page_num}: {e}")

# Criar DataFrame com os dados extraídos
df = pd.DataFrame(data)

# Imprimir resumos e DataFrame
print("Resumos:")
print(resumos)
print("\nDataFrame:")
print(df)
````
